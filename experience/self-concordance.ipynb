```python
# label * (x log(x) - log(x))
def log_x_log_f__(u, label=label):
    return label * (1 + u) / np.power(u, 2)

# - label * log(x)
def log__(u, label=label):
    return label / np.power(u, 2)
```

```python
def w(t):
    return t - np.log(1 + t)

def g(s, x=x, y=y, label=label, f__=f__):
    assert label == 1
    u = s * x + (1 - s) * y
    sfu = np.sqrt(f__(u, label=label))
    return s * w (sfu * np.sqrt(np.power(x - u, 2))) + (1 - s) * w(sfu * np.sqrt(np.power(y - u, 2)))

def g2(s, x=x, y=y, label=label, f__=f__):
    u = s * x + (1 - s) * y
    sfu = np.sqrt(f__(u, label=label))
    return s * (1 - s) * w (sfu * (y - x))

def g3(s, x=x, y=y, label=label, f__=f__):
    sfx = np.sqrt(f__(x, label=label))
    return s * (1 - s) * w(in_w(s, x=x, y=y, label=label, f__=f__))

def g4(s, x=x, y=y, label=label, f__=f__):
    sfx = np.sqrt(f__(x, label=label))
    return s * (1 - s) * w( (y - x) *  sfx * np.sqrt(label) / (np.sqrt(label) + (y - x) * sfx))

def in_w(s, x=x, y=y, label=label, f__=f__):
    sfx = np.sqrt(f__(x, label=label))
    return (y - x) *  sfx * np.sqrt(label) / (np.sqrt(label) + (1-s) * (y - x) * sfx)

def strong_convex(s, x=x, y=y, label=label):
    return 0.5 * s * (1 - s) * label / max(x, y)**2 


#def g4(s):
#    sfx = np.sqrt(f__(x))
#    return s * (1 - s) * w (np.sqrt(sfx) / (1 + y * sfx) * (y - x))

def used_for_log(s, x=x, y=y, label=label, f__=f__):
    return s * (1 - s) * w(np.sqrt(label)) * np.power(x - y, 2) / np.power(y, 2)
```

```python
import matplotlib.pyplot as plt
def fun(y, x):
    r = 1 + np.power(10, y)
    s = np.power(10, x)
    num = np.log(s + (1 - s) * r) - (1 - s) * np.log(r)
    den = s * (1 - s) * (1. / r - 1 + np.log(r))
    return num / den

y = np.linspace(-1, 2, 100)
x = np.linspace(-4, -1,100)

# filling the heatmap, value by value
fun_map = np.empty((x.size, y.size))
for i in range(y.size):
    for j in range(x.size):
        fun_map[i,j] = fun(y[i], x[j])
        
print(fun_map.min(), fun_map.max())

fig = plt.figure()
s = fig.add_subplot(1, 1, 1, xlabel='$s$', ylabel='$r$')
im = s.imshow(
    fun_map,
    extent=(x[0], x[-1], y[0], y[-1]),
    origin='lower', 
    aspect='auto')
fig.colorbar(im)
```
